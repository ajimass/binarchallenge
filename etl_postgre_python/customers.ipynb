## CLEANSING

```python
df_cust_clean = df_customers.copy()
```

```python
df_cust_clean
```

```python
df_cust_clean.columns = map(str.lower, df_cust_clean.columns)
df_cust_clean.columns = df_cust_clean.columns.str.replace(' ', '_')
```

```python
df_cust_clean
```

```python
df_cust_clean = df_cust_clean.rename(columns={'customerkey': 'customer_key'})
```

```python
df_cust_clean
```

```python
# Mengganti nilai NaN dengan 'NAP' pada kolom 'state_code'
df_cust_clean['state_code'] = df_cust_clean['state_code'].fillna('NAP')

print(df_cust_clean)
```

```python
df_cust_clean[df_cust_clean.isna().any(axis=1)]
```

```python
filtered_df = df_cust_clean[df_cust_clean['state'] == 'Napoli']
print(filtered_df)
```

```python
df_cust_clean['birthday'] = pd.to_datetime(df_cust_clean['birthday'])
```

## INGESTION

```python
from sqlalchemy import create_engine
```

```python
connection = 'postgresql://postgres:nunu16@localhost:5432/postgres'
engine = create_engine(connection)
conn2 = engine.connect()
print(conn2)
```

```python
df_cust_clean.to_sql("customers",
                    con=conn2,
                    schema="platinum_base_layer",
                    if_exists="replace",
                    index= False,
                    index_label="index",
                    method="multi")
```

```python
from sqlalchemy import text
```

```python
query_pk = [
"alter table platinum_base_layer.customers add primary key(customer_key)"
]
for query in query_pk:
    conn_2 = engine.connect()
    conn_2.execute(text(query))
    conn_2.commit()
    conn_2.close()
```
