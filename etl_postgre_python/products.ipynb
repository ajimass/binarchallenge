## CLEANSING

```python
df_prod_clean = df_products.copy()
```

```python
df_prod_clean
```

```python
# Menghapus tanda dolar dan mengubah tipe data menjadi float
df_prod_clean['unit_cost_usd'] = df_prod_clean['unit_cost_usd'].str.replace('$','')
df_prod_clean['unit_price_usd'] = df_prod_clean['unit_price_usd'].str.replace('$','')
```

```python
df_prod_clean['unit_cost_usd'] = df_prod_clean['unit_cost_usd'].str.replace(',','').astype(float)
df_prod_clean['unit_price_usd'] = df_prod_clean['unit_price_usd'].str.replace(',','').astype(float)
```

## INGESTION

```python
# engine = create_engine('postgresql://username:password@localhost:5432/mydatabase')
engine = create_engine('postgresql://postgres:nunu16@localhost:5432/postgres')

connection = engine.connect()

df_prod_clean.to_sql( 
        'products', 
        con=engine,
        schema='platinum_base_layer',
        if_exists='replace',    
        index=False,
        index_label = 'index',
        method='multi'
        )
```

```python
from sqlalchemy import text
```

```python
query_pk = [
"alter table platinum_base_layer.products add primary key(product_key)"
]
for query in query_pk:
    conn_2 = engine.connect()
    conn_2.execute(text(query))
    conn_2.commit()
    conn_2.close()
```
